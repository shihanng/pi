# How to get the value?

- So how did we get the value?
- Or more percisely how we get the value of PI?

```python
def genPatches(n):
    circle = plt.Circle((0, 0), 1, fc='y')

    inPoly = patches.RegularPolygon(
        xy=(0, 0),
        numVertices=n,
        radius=1,
        fill=False,
        ec='r',
    )

    b = 1 / math.cos(np.pi/n)
    outPoly = patches.RegularPolygon(
        xy=(0, 0),
        numVertices=n,
        radius=b,
        fill=False,
        ec='b',
    )
    return (circle, inPoly, outPoly)

def inArea(n):
    a = 0.5 * n * np.sin(2*np.pi/n)
    return a

def outArea(n):
    s = 2 * np.tan(np.pi/n)
    a = 0.5 * n * s
    return a

n = 4
circ, inP, outP = genPatches(n)
print(inArea(n))
print(outArea(n))
plt.gca().add_patch(circ); plt.gca().add_patch(inP); plt.gca().add_patch(outP)
plt.axis('equal')
```

```python
plt.show()
```

- The area of the outer square (in blue): $a_\mathrm{out} = (1+1)^2 = 4$

- The area of the inner square (in red): $a_\mathrm{in} = (\sqrt{2})^2 = 2$

- So we know that: $2 \lt a_\mathrm{circ} \lt 4$

- Here is one of the way.
- Remember we want to find the area.
- So we should start from the area of square because it is easier.
- There are two parts to do this.
- They are inner square and outer square.
- Area of inner square is 2
- Area of outer square is 4
- Easy right? But we can do better.

```python
n = 6
circ, inP, outP = genPatches(n)
print(inArea(n))
print(outArea(n))
plt.gca().add_patch(circ); plt.gca().add_patch(inP); plt.gca().add_patch(outP)
plt.axis('equal')
```

```python
plt.show()
```

#### For a hexagon $n=6$:

$$2.598 \lt a_\mathrm{circ} \lt 3.464$$

With some high school trigonometry, we can calculate that the area of the inner and outer polygon as...

```python
n = 8
circ, inP, outP = genPatches(n)
print(inArea(n))
print(outArea(n))
plt.gca().add_patch(circ); plt.gca().add_patch(inP); plt.gca().add_patch(outP)
plt.axis('equal')
```

```python
plt.show()
```

#### For an octagon $n=8$:

$$2.828 \lt a_\mathrm{circ} \lt 3.314$$

```python
n = 10
circ, inP, outP = genPatches(n)
print(inArea(n))
print(outArea(n))
plt.gca().add_patch(circ); plt.gca().add_patch(inP); plt.gca().add_patch(outP)
plt.axis('equal'); 
```

```python
plt.show()
```

#### For a decagon $n=10$:

$$2.939 \lt a_\mathrm{circ} \lt 3.249$$

```python
sides = np.arange(4,50)
inAreas = inArea(sides)
outAreas = outArea(sides)
plt.plot(sides, inAreas, "r-")
plt.plot(sides, outAreas, "b--")
plt.plot(sides, sides*0 + np.pi, "k.")
plt.xlabel("# of sides")
plt.ylabel("area")
```

```python
plt.show()
```

# $a_\mathrm{out}$ (in blue) and $a_\mathrm{in}$ (in red) converge into $\pi$ (in black).

- And if we repeat this process many many times
- we can find that the area of the outter polygon and inner polygon converge into the value of PI.

```python
n = 50
_, inP, outP = genPatches(n)
plt.gca().add_patch(inP); plt.gca().add_patch(outP)
plt.axis('equal'); 
```
